### **Carregamento e preparação dos dados**

```python
# Importando as bibliotecas necessárias
import os
import pandas as pd
```

```python
# Carregando o arquivo CSV em um DataFrame
df = pd.read_csv(r"C:\Users\cassi\GitHub\projeto-excel\Tabelas\vw_olist_full.csv")
```

```python
# Exibindo as informações do DataFrame
df.info()
```

```python
# Exibindo uma amostra dos dados
df.sample(5)
```

```python
# Corrigindo os tipos de dados de colunas de data

date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
    "shipping_limit_date",
    "review_creation_date",
    "review_answer_timestamp"    
]


for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# Criando váriaveis agregadas.
df["entrega_atrasada"] = (df["order_delivered_customer_date"] > df["order_estimated_delivery_date"]).astype(int)
df["tempo_entrega_real"] = (df["order_delivered_customer_date"] - df["order_delivered_carrier_date"]).dt.days
df["tempo_entrega_estimado"] = (df["order_estimated_delivery_date"] - df["order_purchase_timestamp"]).dt.days
```

```python
# Renomeando colunas para nomes mais curtos e intuitivos
df = df.rename(columns={
    "order_purchase_timestamp": "order_purchase_ts",
    "order_approved_at": "order_approved_ts",
    "order_delivered_carrier_date": "carrier_delivery_ts",
    "order_delivered_customer_date": "customer_delivery_ts",
    "order_estimated_delivery_date": "estimated_delivery_ts",
    "customer_zip_code_prefix": "customer_zip_prefix",
    "price": "item_price",
    "product_category_name": "product_category",
    "seller_zip_code_prefix": "seller_zip_prefix",
    "review_comment_title": "review_title",
    "review_comment_message": "review_message",
    "review_creation_date": "review_creation_ts",
    "review_answer_timestamp": "review_answer_ts"
    })
```

### **Métricas Principais (KPIs)**

```python
# --- KPI 1: Taxa de pedidos atrasados ---
taxa_atraso = df['entrega_atrasada'].mean()
# --- KPI 2: Atraso médio em dias ---
df['dif_real_estimado'] = df['tempo_entrega_real'] - df['tempo_entrega_estimado']
atraso_medio = df['dif_real_estimado'].mean()
# --- KPI 3: Tempo médio de entrega real ---
tempo_real_medio = df['tempo_entrega_real'].mean()
# --- KPI 4: Tempo médio estimado ---
tempo_estimado_medio = df['tempo_entrega_estimado'].mean()
# --- KPI 5: Diferença real - estimado ---
erro_estimativa = df['dif_real_estimado'].mean()
```

```python
# Exibir tudo organizado
kpis = {
    "Taxa de atraso (%)": round(taxa_atraso * 100, 2),
    "Atraso médio (dias)": round(atraso_medio, 2),
    "Tempo médio de entrega real (dias)": round(tempo_real_medio, 2),
    "Tempo médio estimado (dias)": round(tempo_estimado_medio, 2),
    "Erro médio da estimativa (dias)": round(erro_estimativa, 2)
}

kpis
```

``“A Olist entrega rápido, mas promete prazos excessivamente longos. Ajustar esse cálculo pode aumentar a conversão e a satisfação.”``

### **Análise por Seller (Proxy de Transportadora)**

```python
# Taxa de atraso por seller
atraso_por_seller = (df.groupby("seller_id")["entrega_atrasada"].mean().sort_values(ascending=False).reset_index())
atraso_por_seller.rename(columns={"entrega_atrasada": "taxa_atraso"}, inplace=True)
atraso_por_seller
```

```python
# Atraso médio em dias por seller (somente pedidos que atrasaram)
atraso_medio_seller = (df[df["entrega_atrasada"] == 1].groupby("seller_id")["tempo_entrega_real"].mean().sort_values(ascending=False).reset_index())
atraso_medio_seller.rename(columns={"tempo_entrega_real": "atraso_medio_dias"}, inplace=True)
atraso_medio_seller
```

```python
# Unir tudo em uma tabela final
seller_metrics = atraso_por_seller.merge(atraso_medio_seller, on="seller_id", how="left")

seller_metrics
```

### **Relação entre atraso e review score**

```python
# Score médio por atraso vs não atraso
score_atraso = df.groupby('entrega_atrasada')['review_score'].mean().reset_index()
score_atraso['entrega_atrasada'] = score_atraso['entrega_atrasada'].map({0: 'No prazo', 1: 'Atrasado'})
score_atraso
```

```python
# Distribuição de notas por atraso
dist_scores = df.groupby(['entrega_atrasada', 'review_score'])['order_id'].count().reset_index()
dist_scores.rename(columns={'order_id': 'qtde'}, inplace=True)
dist_scores
```

```python
# Probabilidade de nota 1 quando há atraso
prob_nota1 = (df[df['review_score'] == 1].groupby('entrega_atrasada')['order_id'].count() / df.groupby('entrega_atrasada')['order_id'].count()).reset_index()

prob_nota1.rename(columns={'order_id': 'prob_nota1'}, inplace=True)
prob_nota1
```

``“Atrasos têm impacto direto e massivo na satisfação.Pedidos atrasados recebem nota 1 quase metade das vezes, puxando o score geral para baixo. A melhoria na logística pode reduzir drasticamente avaliações negativas.”``

### **Correlações e “drivers” do atraso**

```python
# Atraso médio e taxa de atraso por estado
df['customer_state'] = df['customer_state'].astype(str)

atraso_por_estado = (df.groupby('customer_state').agg(taxa_atraso=('entrega_atrasada', 'mean'),atraso_medio_dias=('tempo_entrega_real', 'mean')).reset_index().sort_values('taxa_atraso', ascending=False))

atraso_por_estado
```

```python
# Atraso por peso do produto
atraso_por_peso = (df.groupby(pd.qcut(df['product_weight_g'], 6)).agg(taxa_atraso=('entrega_atrasada', 'mean'),atraso_medio=('tempo_entrega_real', 'mean')).reset_index())

atraso_por_peso.head()
```

```python
# Correlação com variáveis contínuas
cols_continuas = [
    'freight_value', 'product_weight_g',
    'product_length_cm', 'product_height_cm', 'product_width_cm',
    'tempo_entrega_estimado', 'tempo_entrega_real'
]

correlacoes = df[cols_continuas + ['entrega_atrasada']].corr()['entrega_atrasada'].sort_values(ascending=False)
correlacoes
```

```python
# Sellers mais problemáticos
sellers_ruins = ( df.groupby('seller_id').agg(taxa_atraso=('entrega_atrasada', 'mean'),atraso_medio=('tempo_entrega_real', 'mean'),num_pedidos=('seller_id', 'count')).query("num_pedidos >= 20").sort_values('taxa_atraso', ascending=False).head(10))

sellers_ruins
```
